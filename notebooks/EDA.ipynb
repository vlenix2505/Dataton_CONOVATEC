```python
# Convert columns containing "FECHA" to string type
df = df.with_columns([
    pl.col(c).cast(pl.Utf8) for c in df.columns if "FECHA" in c.upper()
])
```

```python
df.schema
```

```python
# Convert FECHA columns to DD/MM/YYYY format (string)
df = df.with_columns([
    pl.when(
        pl.col(c).is_not_null() & (pl.col(c) != "")
    ).then(
        pl.concat_str([
            pl.col(c).str.slice(6, 2),  # DD (offset 6, length 2)
            pl.col(c).str.slice(4, 2),  # MM (offset 4, length 2) 
            pl.col(c).str.slice(0, 4)   # YYYY (offset 0, length 4)
        ], separator="/")
    ).otherwise(None).alias(c)
    for c in df.columns if "FECHA" in c.upper()
])
```

```python
# Convert to timestamp type
df = df.with_columns([
    pl.col(c).str.strptime(pl.Datetime, "%d/%m/%Y").alias(c)
    for c in df.columns if "FECHA" in c.upper()
])
```

```python
import polars as pl

df = df.with_columns([
    pl.when(
        pl.col("NOMBRE_CONSUMO").is_null() & pl.col("PRESENTACION_MEDICAMENTO").is_null()
    ).then(
        None  # Returns null if both are null
    ).otherwise(
        pl.when(
            pl.col("FORMA_FARMACEUTICA").is_not_null()
        ).then(
            pl.concat_str([
                pl.col("NOMBRE_CONSUMO"), 
                pl.col("PRESENTACION_MEDICAMENTO"), 
                pl.col("FORMA_FARMACEUTICA")
            ], separator=" ")
        ).otherwise(
            pl.concat_str([
                pl.col("NOMBRE_CONSUMO"), 
                pl.col("PRESENTACION_MEDICAMENTO")
            ], separator=" ")
        )
    ).alias("CONSUMO_OFICIAL")
])
```

### Outliers

#### Numéricos

```python
import polars as pl

columns_to_check = ["EDAD", "DIAS_HOSP", "CANTIDAD_ENTREGADA", "VALOR_NETO"]

total_rows = df.height

for columna in columns_to_check:
    
    # Calculate quantiles
    quantiles = df.select([
        pl.col(columna).quantile(0.25).alias("Q1"),
        pl.col(columna).quantile(0.75).alias("Q3")
    ]).to_dicts()[0]
    
    Q1, Q3 = quantiles["Q1"], quantiles["Q3"]
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter outlier records
    outliers_df = df.filter(
        (pl.col(columna) < lower_bound) | (pl.col(columna) > upper_bound)
    )

    outliers_count = outliers_df.height
    outliers_pct = round((outliers_count / total_rows) * 100, 2)

    print(f"  Columna: {columna}")
    print(f"  Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
    print(f"  Límite inferior: {lower_bound}, Límite superior: {upper_bound}")
    print(f"  Outliers detectados: {outliers_count} de {total_rows} registros ({outliers_pct}%)")
    print("-"*60)
```

## EDA

```python
import polars as pl
import gc

print("Starting memory-efficient processing...")

# Step 1: Create base aggregations and collect immediately
print("Step 1: Creating base aggregations...")
df_base_agg = (
    df.lazy()
    .group_by('CODIGO_ANONIMIZADO')
    .agg([
        pl.col('SEXO').first(),
        pl.col('EDAD').max().alias('EDAD'),
        pl.col('FECHA_NACIMIENTO').first(),
        pl.col('FECHA_FALLECIMIENTO').first(),
        pl.col('UBIGEO').first(),
        pl.col('DEPARTAMENTO').first(),
        pl.col('PROVINCIA').first(),
        pl.col('DISTRITO').first(),
        pl.col('NIVEL_EESS').first().alias('NIVEL_ESTABLECIMIENTO_PRINCIPAL'),
        pl.col('DIAS_HOSP').sum().alias('TOTAL_DIAS_HOSPITALIZACION'),
        pl.col('VALOR_NETO').sum().alias('TOTAL_GASTO_MEDICAMENTOS'),
        pl.col('FECHA_ATENCION').count().alias('TOTAL_ATENCIONES'),
        pl.col('FECHA_ATENCION').max().alias('ULTIMA_ATENCION'),
        # Binary flags
        (pl.col('FECHA_FALLECIMIENTO').is_not_null().any()).cast(pl.Int8).alias('FALLECIDO'),
        (pl.col('TIPO_DIABETES').is_not_null().any()).cast(pl.Int8).alias('TIENE_DIABETES'),
        (pl.col('DIAS_HOSP') > 0).any().cast(pl.Int8).alias('HOSPITALIZADO_ALGUNA_VEZ')
    ])
    .collect()
)

print(f"Base aggregations completed: {df_base_agg.shape}")
gc.collect()  # Force garbage collection

# Step 2: Diabetes types - collect immediately
print("Step 2: Getting diabetes types...")
df_diabetes_type = (
    df.lazy()
    .filter(pl.col('TIPO_DIABETES').is_not_null())
    .group_by(['CODIGO_ANONIMIZADO', 'TIPO_DIABETES'])
    .agg(pl.len().alias('count'))
    .collect()
)

# Now rank and get top diabetes type
df_diabetes_type = (
    df_diabetes_type.lazy()
    .with_columns(
        pl.col('count').rank(method='dense', descending=True).over('CODIGO_ANONIMIZADO').alias('rank')
    )
    .filter(pl.col('rank') == 1)
    .group_by('CODIGO_ANONIMIZADO')
    .agg(pl.col('TIPO_DIABETES').first().alias('TIPO_DIABETES_PRINCIPAL'))
    .collect()
)

print(f"Diabetes types completed: {df_diabetes_type.shape}")
gc.collect()

# Step 3: Top medications - collect immediately
print("Step 3: Getting top medications...")
df_med_counts = (
    df.lazy()
    .filter(pl.col('CODIGO_CONSUMO').is_not_null())
    .group_by(['CODIGO_ANONIMIZADO', 'CODIGO_CONSUMO', 'NOMBRE_CONSUMO'])
    .agg(pl.col('CANTIDAD_ENTREGADA').sum().alias('TOTAL_CANTIDAD'))
    .collect()
)

# Now rank and get top medication
df_top_med = (
    df_med_counts.lazy()
    .with_columns(
        pl.col('TOTAL_CANTIDAD').rank(method='dense', descending=True).over('CODIGO_ANONIMIZADO').alias('rank')
    )
    .filter(pl.col('rank') == 1)
    .group_by('CODIGO_ANONIMIZADO')
    .agg([
        pl.col('CODIGO_CONSUMO').first().alias('PRINCIPAL_CODIGO_CONSUMO'),
        pl.col('NOMBRE_CONSUMO').first().alias('PRINCIPAL_NOMBRE_CONSUMO')
    ])
    .collect()
)

print(f"Top medications completed: {df_top_med.shape}")

# Clean up intermediate dataframes
del df_med_counts
gc.collect()

# Step 4: Final joins
print("Step 4: Joining everything...")
df_consolidado = (
    df_base_agg.lazy()
    .join(df_diabetes_type.lazy(), on='CODIGO_ANONIMIZADO', how='left')
    .join(df_top_med.lazy(), on='CODIGO_ANONIMIZADO', how='left')
    .sort('CODIGO_ANONIMIZADO')
    .collect()
)

# Clean up intermediate dataframes
del df_base_agg, df_diabetes_type, df_top_med
gc.collect()

print("✅ Completed!")
print(f"Final shape: {df_consolidado.shape}")
print("\nSchema:")
print(df_consolidado.schema)
print("\nFirst 5 rows:")
print(df_consolidado.head())
```

```python
import polars as pl
import matplotlib.pyplot as plt

# Agrupar y contar
counts_df = (
    df_consolidado
    .group_by("TIPO_DIABETES_PRINCIPAL")
    .agg(pl.len().alias("count"))
    .sort("count", descending=True)
    .head(10)
)

# Convertir a Pandas
counts_pd = counts_df.to_pandas().dropna()
counts_pd["percentage"] = (counts_pd["count"] / counts_pd["count"].sum()) * 100

# Gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.barh(
    counts_pd["TIPO_DIABETES_PRINCIPAL"],
    counts_pd["count"],
    color=plt.cm.Pastel1.colors
)

plt.xlabel("Cantidad")
plt.ylabel("Tipo de Diabetes Principal")
plt.title("Top 10 TIPO_DIABETES_PRINCIPAL (por cantidad)")

# Invertir el eje Y para que el más grande quede arriba
plt.gca().invert_yaxis()

# Mostrar etiquetas con el porcentaje en cada barra
for bar, pct in zip(bars, counts_pd["percentage"]):
    plt.text(
        bar.get_width() + max(counts_pd["count"]) * 0.01,  # un poco a la derecha del final de la barra
        bar.get_y() + bar.get_height()/2,
        f"{pct:.1f}%",
        va='center'
    )

plt.tight_layout()
plt.show()
```

##### Análisis bivariado

```python
import polars as pl

# Check if FECHA_NACIMIENTO is already a datetime or needs conversion
print("Checking FECHA_NACIMIENTO format...")
sample_dates = df_consolidado.select("FECHA_NACIMIENTO").head(5)
print(sample_dates)

# If it's already datetime, we can work directly with it
df_consolidado = df_consolidado.with_columns([
    # If FECHA_NACIMIENTO is already datetime, extract date part
    pl.col("FECHA_NACIMIENTO").dt.date().alias("FECHA_NACIMIENTO_DATE"),
    
    # Extract year directly
    pl.col("FECHA_NACIMIENTO").dt.year().alias("ANIO_NAC"),
    
    # Classify age groups
    pl.when(pl.col("EDAD") < 18)
      .then(pl.lit("Menor de edad"))
      .when((pl.col("EDAD") >= 18) & (pl.col("EDAD") <= 39))
      .then(pl.lit("Joven-adulto"))
      .when((pl.col("EDAD") >= 40) & (pl.col("EDAD") <= 64))
      .then(pl.lit("Adulto"))
      .otherwise(pl.lit("Adulto mayor"))
      .alias("GRUPO_ETARIO")
])
```

```python
import polars as pl

result = (df_consolidado
    .group_by(["SEXO", "GRUPO_ETARIO"])
    .agg(pl.len().alias("count"))
    .sort(["SEXO", "GRUPO_ETARIO"])
)

print(result)
```

```python
import polars as pl

result = (df_consolidado
    .group_by(["TIPO_DIABETES_PRINCIPAL", "GRUPO_ETARIO"])
    .agg(pl.len().alias("count"))
    .sort(["TIPO_DIABETES_PRINCIPAL", "GRUPO_ETARIO"])
)

print(result)
```

```python
import polars as pl

result = (
    df_consolidado
    .group_by("SEXO")
    .agg([
        pl.col("FALLECIDO").mean().alias("TASA_MORTALIDAD"),
        pl.col("FALLECIDO").sum().alias("TOTAL_FALLECIDOS")
    ])
    .sort("SEXO")
)

print(result)
```

```python
import polars as pl
import matplotlib.pyplot as plt

# Procesamiento con Polars
sexo_df = (df_consolidado
    .group_by("SEXO")
    .agg(pl.col("FALLECIDO").mean().alias("TASA_MORTALIDAD"))
    .sort("SEXO")
    .to_pandas()  # Convertir a pandas solo para el gráfico
)

# Gráfico de barras
sexo_df.plot(kind='bar', x='SEXO', y='TASA_MORTALIDAD', legend=False, color='skyblue')
plt.ylabel("Tasa de Mortalidad")
plt.title("Mortalidad por Sexo")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import polars as pl

result = (
    df_consolidado
    .group_by("TIPO_DIABETES_PRINCIPAL")
    .agg([
        pl.col("FALLECIDO").mean().alias("TASA_MORTALIDAD"),
        pl.col("FALLECIDO").sum().alias("TOTAL_FALLECIDOS"),
        pl.col("FALLECIDO").count().alias("TOTAL_REGISTROS")
    ])
    .sort("TIPO_DIABETES_PRINCIPAL")
)

print(result)
```

```python
import polars as pl
import matplotlib.pyplot as plt

# Calcular tasa de mortalidad por tipo de diabetes con Polars
diabetes_df = (df_consolidado
    .group_by("TIPO_DIABETES_PRINCIPAL")
    .agg(pl.col("FALLECIDO").mean().alias("TASA_MORTALIDAD"))
    .sort("TASA_MORTALIDAD", descending=True)  # Ordenar por tasa para mejor visualización
    .to_pandas()  # Convertir a pandas solo para el gráfico
)

# Gráfico de barras
diabetes_df.plot(kind='bar', x='TIPO_DIABETES_PRINCIPAL', y='TASA_MORTALIDAD', legend=False, color='salmon')
plt.ylabel("Tasa de Mortalidad")
plt.title("Mortalidad por Tipo de Diabetes")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
